```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from sklearn.cluster import KMeans
```

```python
# Generate some numbers
data = np.random.randn(400,2)
data[:100] += 5, 5
data[100:200] += 10, 10
data[200:300] += 10, 5
data[300:] += 5, 10
```

```python
sns.scatterplot(x = data[:, 0], y =data[:, 1], data = data);
```

```python
model = KMeans(n_clusters=4, init='random', random_state=42, max_iter=10, n_init=1)
model.fit(data)
```

```python
y_pred = model.predict(data)
```

```python
plt.scatter(x = data[:, 0], y =data[:, 1], c = y_pred)
plt.scatter(x = model.cluster_centers_[:,0], y = model.cluster_centers_[:, 1], c= 'r')
plt.show()
```

```python
# Kmeans with n_cluster =8
model = KMeans(n_clusters=8, init='random', random_state=42, max_iter=10, n_init=1)
model.fit(data)
y_pred = model.predict(data)

#Viz results
plt.scatter(x = data[:, 0], y =data[:, 1], c = y_pred)
plt.scatter(x = model.cluster_centers_[:,0], y = model.cluster_centers_[:, 1], c= 'r')
plt.show()
```
